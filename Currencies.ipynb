```python
import requests
import xmltodict
url = "http://www.cbr.ru/scripts/XML_val.asp"
response = requests.get(url)
response
```

```python
response.content
```

```python
data = xmltodict.parse(response.content)
data
```

```python
items = []
for item in data['Valuta']['Item']:
    items.append(item)
items
```

```python
my_array = []
for item in data['Valuta']['Item']:
    my_set=[item['EngName'],item['Name'], item['Nominal']]
    my_array.append(my_set)
    print(my_set)
```
